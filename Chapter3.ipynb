# Visualizing Data

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [300.2, 543.3, 1075.9, 2862.5, 5979.6, 10289.7, 14958.3]
```

```python
plt.plot(years, gdp, color='green', marker='o')

plt.ylabel('Billion of $')
plt.title('Norminal GDP')
plt.show()
```

## Line Charts

```python
variance     = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = [256, 128, 64, 32, 16, 8, 4, 2, 1]
total_error  = [x + y for x, y in zip(variance, bias_squared)]
```

```python
xs = range(len(variance))
```

```python
plt.plot(xs, variance, 'g-', label='variance')
plt.plot(xs, bias_squared, 'r-.', label='bias')
plt.plot(xs, total_error, 'b:', label='total_error')
plt.legend()
plt.title('The Bias-Variance Tradeoff')
plt.xlabel('Model Complexity')
plt.xticks([])
plt.show()
```

## Scatter Plot

```python
friends = [ 70,  65,  72,  63,  71,  64,  60,  64,  67]
minutes = [175, 170, 205, 120, 220, 130, 105, 145, 190]
labels =  ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i']
```

```python
plt.scatter(friends, minutes)
for label, fcnt, mins in zip(labels, friends, minutes):
    plt.annotate(label, 
                 xy = (fcnt, mins), 
                 xytext=(5,-5),
                 textcoords='offset points'
                )
plt.xlabel('# of friends')
plt.ylabel('daily spend minutes on the site')
plt.title('Daily Minutes vs # of Friends')
plt.show()
```
